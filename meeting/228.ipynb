```python
import codecs
import os
from ipywidgets import IntProgress
from IPython.display import display

modalverb = ["應","要","可","能","可以","須","應該","必須","會","得","需要","當","應當","能夠","該","需"]
path = "C:\\Users\\user\\Desktop\\SNA\\meeting\\source\\228\\"
out_path = "C:\\Users\\user\\Desktop\\"

file_list = []
count = 0

for file in os.listdir(path):
    file_list.append(file)
    
p = IntProgress()
p.max = len(file_list)
p.description = 'start'
display(p)
    
dic = {}
sen = {}

for file in file_list:
    with codecs.open(path+file,'rb','utf8') as f:
        contents = f.readlines()

        for i in contents:
            
            try:

                temp = i.strip().split()

                index = 0
                pairs = []

                for j in temp:
                    words = j.split('(')
                    words[1] = '(' + words[1]

                    if words[0] in modalverb:
                        if index != len(temp)-1:
                            if (words[1] == '(D)' or words[1][1] == 'V') and \
                                (temp[index+1].split('(')[1][0] == 'V' or temp[index+1].split('(')[1][0] == 'D' \
                                or temp[index+1].split('(')[1][0] == 'P' or temp[index+1].split('(')[1][0] == 'C'):
                                pairs.append(temp[index])
                    index += 1

                if len(pairs) > 1:

                    for j in range(len(pairs)-1):    
                        term = pairs[j]+'_'+pairs[j+1]

                        if term not in dic:
                            dic[term] = 1
                        else:
                            dic[term] += 1

                        if term not in sen:
                            sen[term] = [''.join(temp)]
                        elif len(sen[term]) < 3:
                            sen[term].append(''.join(temp))
            except:
                with codecs.open("C:\\Users\\user\\Desktop\\error.txt",'ab','utf8') as g:
                    try:
                        g.write(file+'\t'+i+'\r\n')
                    except:
                        g.write(file+'\tNone\r\n')
                        
    count = count + 1
    p.value = count
    p.description = str(count)

dic = sorted(dic.items(), key=lambda d:d[1], reverse = True)
with codecs.open(out_path+'output.txt','wb','utf8') as f:
    for i in dic:
        #print (i[0],i[1])
        f.write(str(i[0])+' '+str(i[1]))
        '''for j in sen[i[0]]:
            f.write(j+'\r\n')'''
        f.write('\r\n')

    
p.description = 'end'
```
