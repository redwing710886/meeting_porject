```python
import codecs
import time
from ipywidgets import IntProgress
from IPython.display import display

path = "C:\\Users\\user\\Desktop\\課業相關\\碩士班\\SNA\\meeting\\人稱代名詞\\source\\"
out_path = "C:\\Users\\user\\Desktop\\"
node_id = {}
all_chain = []
id_count = 1
all_count = 0

with codecs.open(path+'我.txt','rb','utf8') as f:
    content = f.readlines()
    
    p = IntProgress()
    p.max = len(content)
    p.description = 'start'
    display(p)
    
    for i in content:
        term = i.strip().split()
        words = ''
        link = []
        
        for j in term:
            word = j.split('(')[0]
            if word == '':
                word = '('
                
            if word not in node_id:
                node_id[word] = id_count
                link.append(id_count)
                id_count = id_count + 1
            else:
                link.append(node_id[word])
                
            words = words + ' ' + word
        
        #print (words)
        
        chain = []
        for count in range(len(link)-1):
            chain.append(str(link[count])+'->'+str(link[count+1]))
            all_chain.append(str(link[count])+'->'+str(link[count+1]))
        #print (' '.join(chain))
        
        all_count = all_count + 1
        p.value = all_count
        p.description = str(all_count)
        
        #if all_count > 500:
        #    break
        
        #time.sleep(0.3)
        
with codecs.open(out_path+'node.txt','wb','utf8') as f:
    f.write("id\tname\tweight\n")
    
    node = sorted(node_id.items(), key=lambda d:d[1], reverse = False)
    
    for i in node:
        f.write(str(i[1])+'\t'+str(i[0])+'\t1\n')

with codecs.open(out_path+'link.txt','wb','utf8') as f:
    f.write("source\ttarget\ttype\tweight\n")
    
    for i in all_chain:
        f.write(i.split('->')[0]+'\t'+i.split('->')[1]+'\tdirected\t1\n')
        
        
p.description = 'end'
```
